```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import butter, filtfilt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix,
                              accuracy_score, f1_score)
import lightgbm as lgb
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import warnings
warnings.filterwarnings('ignore')

SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

ACTIVITIES = {
    1: 'Walking',
    2: 'Walking_Upstairs',
    3: 'Walking_Downstairs',
    4: 'Sitting',
    5: 'Standing',
    6: 'Laying'
}
act_labels = [ACTIVITIES[i+1] for i in range(6)]
```

```python
BASE = 'UCI HAR Dataset'

def load_raw_signals(split):
    sig_path = f'{BASE}/{split}/Inertial Signals/'
    files = [
        'body_acc_x', 'body_acc_y', 'body_acc_z',
        'body_gyro_x', 'body_gyro_y', 'body_gyro_z',
        'total_acc_x', 'total_acc_y', 'total_acc_z'
    ]
    signals = []
    for f in files:
        data = np.loadtxt(f'{sig_path}{f}_{split}.txt')
        signals.append(data)
    return np.stack(signals, axis=2)   # (n_samples, 128, 9)

def load_labels(split):
    y = np.loadtxt(f'{BASE}/{split}/y_{split}.txt', dtype=int)
    return y - 1   # 0-indexed

def load_features(split):
    return np.loadtxt(f'{BASE}/{split}/X_{split}.txt')   # (n_samples, 561)

X_raw_train  = load_raw_signals('train')    # (7352, 128, 9)
X_raw_test   = load_raw_signals('test')     # (2947, 128, 9)
y_train      = load_labels('train')
y_test       = load_labels('test')
X_feat_train = load_features('train')       # (7352, 561)
X_feat_test  = load_features('test')        # (2947, 561)

print(f"Raw train shape    : {X_raw_train.shape}")
print(f"Raw test shape     : {X_raw_test.shape}")
print(f"Feature train shape: {X_feat_train.shape}")
print(f"Classes            : {np.unique(y_train)}")
```

```python
def butter_lowpass(cutoff=20, fs=50, order=4):
    nyq = fs / 2
    b, a = butter(order, cutoff / nyq, btype='low')
    return b, a

def apply_filter(signal_3d):
    """Apply Butterworth low-pass filter to first 6 channels (accel + gyro)."""
    b, a = butter_lowpass()
    filtered = signal_3d.copy()
    for i in range(signal_3d.shape[0]):
        for ch in range(6):
            filtered[i, :, ch] = filtfilt(b, a, signal_3d[i, :, ch])
    return filtered

def normalize_signals(X_train, X_test):
    """Normalize signals to [-1, 1] using training statistics."""
    n_ch = X_train.shape[2]
    X_tr_flat = X_train.reshape(-1, n_ch)
    mins = X_tr_flat.min(axis=0)
    maxs = X_tr_flat.max(axis=0)
    def scale(X):
        return 2 * (X - mins) / (maxs - mins + 1e-8) - 1
    return (scale(X_train.reshape(-1, n_ch)).reshape(X_train.shape),
            scale(X_test.reshape(-1, n_ch)).reshape(X_test.shape))

print("Applying Butterworth low-pass filter (20 Hz cutoff)...")
X_raw_train_f = apply_filter(X_raw_train[:, :, :6])
X_raw_test_f  = apply_filter(X_raw_test[:, :, :6])

X_raw_train_n, X_raw_test_n = normalize_signals(X_raw_train_f, X_raw_test_f)

print(f"Filtered & normalized. Value range: "
      f"[{X_raw_train_n.min():.2f}, {X_raw_train_n.max():.2f}]")
```

```python
scaler = StandardScaler()
X_feat_train_s = scaler.fit_transform(X_feat_train)
X_feat_test_s  = scaler.transform(X_feat_test)

pca = PCA(n_components=50, random_state=SEED)
X_pca_train = pca.fit_transform(X_feat_train_s)
X_pca_test  = pca.transform(X_feat_test_s)

print(f"PCA explained variance (50 PCs): "
      f"{pca.explained_variance_ratio_.sum()*100:.1f}%")

# Plot cumulative explained variance
plt.figure(figsize=(8, 3))
plt.plot(np.cumsum(pca.explained_variance_ratio_) * 100, color='#185FA5', lw=2)
plt.axhline(y=pca.explained_variance_ratio_.cumsum()[-1]*100,
            ls='--', color='gray', lw=0.8)
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance (%)')
plt.title('PCA — Cumulative Explained Variance')
plt.tight_layout()
plt.show()
```

```python
def plot_confusion(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, normalize='true')
    fig, ax = plt.subplots(figsize=(7, 5))
    sns.heatmap(cm * 100, annot=True, fmt='.0f', cmap='Blues',
                xticklabels=[a[:4] for a in act_labels],
                yticklabels=[a[:4] for a in act_labels], ax=ax)
    ax.set_title(title)
    ax.set_ylabel('True label')
    ax.set_xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

def print_metrics(y_true, y_pred, name):
    acc = accuracy_score(y_true, y_pred) * 100
    f1  = f1_score(y_true, y_pred, average='macro') * 100
    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"  Overall Accuracy : {acc:.2f}%")
    print(f"  Macro F1-Score   : {f1:.2f}%")
    print(f"{'='*55}")
    print(classification_report(y_true, y_pred, target_names=act_labels))
    return acc, f1

print("Helper functions defined.")
```

```python
print("Training LightGBM...")

lgb_params = dict(
    objective='multiclass',
    num_class=6,
    n_estimators=500,
    learning_rate=0.05,
    num_leaves=63,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=SEED,
    n_jobs=-1,
    verbose=-1
)

lgb_model = lgb.LGBMClassifier(**lgb_params)
lgb_model.fit(
    X_pca_train, y_train,
    eval_set=[(X_pca_test, y_test)],
    callbacks=[lgb.early_stopping(50, verbose=False)]
)

y_pred_lgb = lgb_model.predict(X_pca_test)
acc_lgb, f1_lgb = print_metrics(y_test, y_pred_lgb, "LightGBM (PCA-50)")
plot_confusion(y_test, y_pred_lgb, "LightGBM — Confusion Matrix")

import os
lgb_model.booster_.save_model('/tmp/lgb_model.txt')
print(f"Model size: {os.path.getsize('/tmp/lgb_model.txt')/1024:.0f} KB")
```

```python
class CNN_LSTM(nn.Module):
    def __init__(self, n_channels=6, n_classes=6):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv1d(n_channels, 64, kernel_size=5, padding=2),
            nn.BatchNorm1d(64), nn.ReLU(),
            nn.Conv1d(64, 64, kernel_size=5, padding=2),
            nn.BatchNorm1d(64), nn.ReLU(),
            nn.MaxPool1d(2), nn.Dropout(0.3)
        )
        self.conv2 = nn.Sequential(
            nn.Conv1d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm1d(128), nn.ReLU(),
            nn.MaxPool1d(2), nn.Dropout(0.3)
        )
        self.lstm = nn.LSTM(128, 128, num_layers=2,
                            batch_first=True, dropout=0.3)
        self.classifier = nn.Sequential(
            nn.Linear(128, 64), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(64, n_classes)
        )

    def forward(self, x):           # x: (B, T, C)
        x = x.permute(0, 2, 1)     # → (B, C, T) for Conv1d
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.permute(0, 2, 1)     # → (B, T', C') for LSTM
        out, _ = self.lstm(x)
        x = out[:, -1, :]          # last timestep
        return self.classifier(x)

print("CNN_LSTM architecture defined.")
print(f"Parameters: {sum(p.numel() for p in CNN_LSTM().parameters()):,}")
```

```python
def make_loaders(X_tr, y_tr, X_te, y_te, batch_size=128):
    Xtr = torch.tensor(X_tr, dtype=torch.float32)
    ytr = torch.tensor(y_tr, dtype=torch.long)
    Xte = torch.tensor(X_te, dtype=torch.float32)
    yte = torch.tensor(y_te, dtype=torch.long)
    tr_loader = DataLoader(TensorDataset(Xtr, ytr),
                           batch_size=batch_size, shuffle=True)
    te_loader = DataLoader(TensorDataset(Xte, yte),
                           batch_size=batch_size)
    return tr_loader, te_loader

def train_model(model, tr_loader, te_loader, epochs=30, lr=1e-3):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, epochs)
    criterion = nn.CrossEntropyLoss()
    history   = {'tr_acc': [], 'te_acc': [], 'tr_loss': [], 'te_loss': []}

    for ep in range(1, epochs + 1):
        # ── train ──
        model.train()
        tr_loss, tr_correct, n = 0.0, 0, 0
        for xb, yb in tr_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            out  = model(xb)
            loss = criterion(out, yb)
            loss.backward()
            optimizer.step()
            tr_loss    += loss.item() * len(yb)
            tr_correct += (out.argmax(1) == yb).sum().item()
            n          += len(yb)
        scheduler.step()

        # ── evaluate ──
        model.eval()
        te_loss, te_correct, nt = 0.0, 0, 0
        with torch.no_grad():
            for xb, yb in te_loader:
                xb, yb = xb.to(device), yb.to(device)
                out  = model(xb)
                loss = criterion(out, yb)
                te_loss    += loss.item() * len(yb)
                te_correct += (out.argmax(1) == yb).sum().item()
                nt         += len(yb)

        history['tr_acc'].append(tr_correct / n * 100)
        history['te_acc'].append(te_correct / nt * 100)
        history['tr_loss'].append(tr_loss / n)
        history['te_loss'].append(te_loss / nt)

        if ep % 5 == 0 or ep == 1:
            print(f"Epoch {ep:02d}/{epochs}  |  "
                  f"Train {tr_correct/n*100:.1f}%  |  "
                  f"Test  {te_correct/nt*100:.1f}%")

    return history

def predict_all(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for xb, _ in loader:
            preds.append(model(xb.to(device)).argmax(1).cpu().numpy())
    return np.concatenate(preds)

print("Training utilities defined.")
```

```python
print("Training 1D-CNN + LSTM...")
tr_loader, te_loader = make_loaders(X_raw_train_n, y_train,
                                    X_raw_test_n,  y_test)

cnn_lstm = CNN_LSTM().to(device)
hist_cnn = train_model(cnn_lstm, tr_loader, te_loader, epochs=30, lr=1e-3)

y_pred_cnn = predict_all(cnn_lstm, te_loader)
acc_cnn, f1_cnn = print_metrics(y_test, y_pred_cnn, "1D-CNN + LSTM")
plot_confusion(y_test, y_pred_cnn, "1D-CNN + LSTM — Confusion Matrix")

# Training curves
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(hist_cnn['tr_acc'], label='Train')
axes[0].plot(hist_cnn['te_acc'], label='Test')
axes[0].set_title('CNN-LSTM — Accuracy')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy (%)'); axes[0].legend()

axes[1].plot(hist_cnn['tr_loss'], label='Train')
axes[1].plot(hist_cnn['te_loss'], label='Test')
axes[1].set_title('CNN-LSTM — Loss')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss'); axes[1].legend()
plt.tight_layout(); plt.show()
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=128, dropout=0.1):
        super().__init__()
        self.drop = nn.Dropout(dropout)
        pe  = torch.zeros(max_len, d_model)
        pos = torch.arange(max_len).unsqueeze(1).float()
        div = torch.exp(torch.arange(0, d_model, 2).float()
                        * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(pos * div)
        pe[:, 1::2] = torch.cos(pos * div)
        self.register_buffer('pe', pe.unsqueeze(0))   # (1, max_len, d_model)

    def forward(self, x):
        return self.drop(x + self.pe[:, :x.size(1)])

class IMUTransformer(nn.Module):
    def __init__(self, n_channels=6, d_model=64, nhead=4,
                 num_layers=3, n_classes=6, dropout=0.2):
        super().__init__()
        self.input_proj = nn.Linear(n_channels, d_model)
        self.pos_enc    = PositionalEncoding(d_model, dropout=dropout)
        enc_layer = nn.TransformerEncoderLayer(
            d_model, nhead,
            dim_feedforward=128,
            dropout=dropout,
            batch_first=True
        )
        self.encoder    = nn.TransformerEncoder(enc_layer, num_layers)
        self.classifier = nn.Sequential(
            nn.LayerNorm(d_model),
            nn.Linear(d_model, 32), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(32, n_classes)
        )

    def forward(self, x):           # x: (B, T, C)
        x = self.input_proj(x)
        x = self.pos_enc(x)
        x = self.encoder(x)
        x = x.mean(dim=1)          # global average pooling over time
        return self.classifier(x)

print("IMUTransformer architecture defined.")
print(f"Parameters: {sum(p.numel() for p in IMUTransformer().parameters()):,}")
```

```python
print("Training Transformer...")
tfm_model = IMUTransformer().to(device)
hist_tfm  = train_model(tfm_model, tr_loader, te_loader, epochs=30, lr=5e-4)

y_pred_tfm = predict_all(tfm_model, te_loader)
acc_tfm, f1_tfm = print_metrics(y_test, y_pred_tfm, "Transformer")
plot_confusion(y_test, y_pred_tfm, "Transformer — Confusion Matrix")

# Training curves
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(hist_tfm['tr_acc'], label='Train')
axes[0].plot(hist_tfm['te_acc'], label='Test')
axes[0].set_title('Transformer — Accuracy')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Accuracy (%)'); axes[0].legend()

axes[1].plot(hist_tfm['tr_loss'], label='Train')
axes[1].plot(hist_tfm['te_loss'], label='Test')
axes[1].set_title('Transformer — Loss')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss'); axes[1].legend()
plt.tight_layout(); plt.show()
```
